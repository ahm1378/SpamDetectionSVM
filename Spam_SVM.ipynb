```python
X = df.drop("spam", axis = 1)
y = df.spam.values.astype(int)
```

```python
X.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform (X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
p = svc.predict(X_train)
```

```python
accuracy = np.mean(p==y_train)
print("accuracy",svc.score(X_train,y_train))
```

the default kenerl is rbf

```python
c_range=[0.01,0.1,1,10,100,1000]
for c in c_range:
    rbf = SVC(kernel = 'rbf', random_state = 0,C=c)
    rbf.fit(X_train, y_train)
    accuracy_train = rbf.score(X_train,y_train)
    accuracy_test = rbf.score(X_test,y_test)
    print("accuracy_train is  {} , accuracy_test is  {}, c  is {}".format(accuracy_train,accuracy_test,c))
```

```python
c_range=[0.01,0.1,1,10,100]
for c in c_range:
    rbf = SVC(kernel = 'linear', random_state = 0,C=c)
    rbf.fit(X_train, y_train)
    accuracy_train = rbf.score(X_train,y_train)
    accuracy_test = rbf.score(X_test,y_test)
    print("accuracy_train is  {} , accuracy_test is  {}, c  is {}".format(accuracy_train,accuracy_test,c))
```

```python
c_range=[0.01,0.1,1,10,100]
for c in c_range:
    rbf = SVC(kernel = 'poly', random_state = 0,C=c)
    rbf.fit(X_train, y_train)
    accuracy_train = rbf.score(X_train,y_train)
    accuracy_test = rbf.score(X_test,y_test)
    print("accuracy_train is  {} , accuracy_test is  {}, c  is {}".format(accuracy_train,accuracy_test,c))
```

the best c for rbf is 10 ,
the best c for linear is 100,
